##**Data Cleaning**

##**Business Requirements**

####**Kp's Requirements**

```python
#Total Sales
Total_sales = df['Sales'].sum()

#Average Sales
Avg_sales = df['Sales'].mean()

#no of Item Sold
No_of_Items_sold = df['Sales'].count()

#Avg_Rating
Avg_Rating = df['Rating'].mean()

#Display
print(f"Total_Sales :${Total_sales:,.0f}")
print(f"Avg_sales :${Avg_sales:,.0f}")
print(f"No_of_Items_sold :${No_of_Items_sold:,.0f}")
print(f"Avg_Rating :${Avg_Rating:,.0f}")
```

##**Charts Requirements**

Total Sales

```python
Sales_by_fat = df.groupby('Item Fat Content')['Sales'].sum()
plt.pie(Sales_by_fat,labels= Sales_by_fat.index,
        autopct='%1.1f%%',
        startangle=90 )
plt.title('Sales By Fat Content')
plt.axis('equal')
plt.axis('tight')
plt.show()
```

Total Sales By item Type


```python
Sales_by_Type = df.groupby('Item Type')['Sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(10,6))
bars=plt.bar(Sales_by_Type.index,Sales_by_Type.index)
plt.xticks(rotation=-90)
plt.title('Sales By Item Type')
plt.xlabel('Item Type')
plt.ylabel('Total Sales')
plt.title('Total Sales by item type')

for bar in bars:
  plt.text(bar.get_x()+bar.get_width()/2,bar.get_height(),
           f'${bar.get_height():,.0f}',ha='center',va='bottom',fontsize=8)
plt.tight_layout()
plt.show()
```

fat content by outlet for total sales

```python
# Group and reshape data
grouped = df.groupby(['Outlet Location Type', 'Item Fat Content'])['Sales'].sum().unstack()

# Ensure columns are in the desired order and exist
grouped = grouped[['Regular', 'Low Fat']]

# Plot
ax = grouped.plot(kind='bar', figsize=(8, 5), title='Outlet tier by item Fat Content')

# Set axis labels correctly
ax.set_xlabel('Outlet Location')
ax.set_ylabel('Total Sales')

# Set legend title
ax.legend(title='Item Fat Content')

# Adjust layout and display plot
plt.tight_layout()  # make sure plt is still matplotlib.pyplot
plt.show()
```

Total Sales by outlet Establishment

```python
Sales_by_year = df.groupby('Outlet Establishment Year')['Sales'].sum()
plt.figure(figsize=(10,6))
plt.plot(Sales_by_year.index,Sales_by_year.values,marker='o',linestyle='-')
plt.xticks(rotation=-90)
plt.xlabel('Outlet Establishment Year')
plt.ylabel('Total Sales')
plt.title('Outlet Establishment')
for x,y in zip(Sales_by_year.index,Sales_by_year.values):
  plt.text(x,y,f'${y:,.0f}',ha='center',va='bottom')
plt.tight_layout()
plt.show()
```

Sales by Outelt size

```python
Sales_by_size = df.groupby('Outlet Size')['Sales'].sum()
plt.figure(figsize=(10,6))
plt.pie(Sales_by_size,labels=Sales_by_size.index,
                          autopct='%1.1f%%',
                          startangle=90)
plt.title('Sales by oultlet Size')
plt.tight_layout()
plt.show()
```

```python
numerical_df = df.select_dtypes(include=['number'])

# 4. Compute the correlation matrix
correlation_matrix = numerical_df.corr()

# 5. Plot the heatmap using seaborn
plt.figure(figsize=(10, 8))  # Set the size of the plot
sns.heatmap(correlation_matrix,
            annot=True,      # Display correlation values
            cmap='coolwarm', # Color theme
            fmt=".2f",       # Format of correlation numbers
            linewidths=0.5,  # Line between boxes
            square=True)     # Make the heatmap square

# 6. Add title and display
plt.title('Correlation Heatmap of Numeric Features')
plt.tight_layout()
plt.show()
```
